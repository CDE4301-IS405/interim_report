```python
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data_2ml = pd.read_excel('Pilot study results.xlsx',
                     sheet_name='cleaned sheet (run 1)',
                     skiprows=2, usecols='B:AK')
Data_2ml
```

```python
time = Data_2ml['Time [h]']
```

```python
VOC = Data_2ml.filter(like='VOC').copy()
VOC['Time'] = time
VOC.dropna()
VOC = VOC.drop(['VOC level (control 2)'], axis = 1)
VOC.columns = ['empty','control','plant','plant+microbe','Time']
```

```python
import plotly.graph_objects as go

# Fixed colors
colors = {
    'empty': 'blue',
    'control': 'orange',
    'plant': 'turquoise',
    'plant+microbe': 'green'
}

series_names = ['empty', 'control', 'plant', 'plant+microbe']

# Create figure
fig = go.Figure()

for col in series_names:
    fig.add_trace(go.Scatter(
        x=VOC['Time'],
        y=VOC[col],
        mode='lines+markers',
        name=col.capitalize(),
        line=dict(color=colors[col]),
        marker=dict(color=colors[col]),
        visible=True  # all visible initially
    ))

# Layout with subtitle directly below the title
fig.update_layout(
    title=dict(
        text="VOC Concentration Over Time [2.5ml of each pollutant] <br><sup>Click on the series in the legend to include/exclude for better comparison</sup>",
        x=0.05,  
        xanchor='left'
    ),
    xaxis_title='Time (h)',
    yaxis_title='VOC (ppb)',
    legend=dict(title='Series', x=1.05, y=1)
)

fig.show()
# Export interactive figure as HTML
fig.write_html("VOC_Concentration_Interactive.html", include_plotlyjs='cdn', full_html=True)
```
